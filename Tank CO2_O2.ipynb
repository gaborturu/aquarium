```python
tank = Tank(volume = 100,
            headspace = 20,
            KH=10,
            room_CO2_equilibrium=1
            )

CO2 = []
O2 = []

for o2 in range(1000):
    tank.use_O2(1, 0.7)
    CO2.append(tank.CO2_water_concentration)
    O2.append(tank.O2_water_concentration)

fig = plt.figure()
plt.plot(O2, label="O2 concentration")
plt.plot(CO2, label="CO2 concentration")
plt.legend()
plt.xlabel("O2 used (mg)")
plt.ylabel("Gas concentration in water (mg/l)")
plt.title("20 L headspace")
sns.despine()
plt.show()

fig.savefig("headspace_CO2_O2.png")
```

```python
tank = Tank(volume = 120,
            headspace = 0,
            KH=10,
            room_CO2_equilibrium=1
            )

CO2 = []
O2 = []

for o2 in range(1000):
    tank.use_O2(1, 0.7)
    CO2.append(tank.CO2_water_concentration)
    O2.append(tank.O2_water_concentration)

fig = plt.figure()
plt.plot(O2, label="O2 concentration")
plt.plot(CO2, label="CO2 concentration")
plt.legend()
plt.xlabel("used O2 (mg)")
plt.ylabel("Gas concentration in water (mg/l)")
plt.title("no headspace")
sns.despine()
plt.show()
fig.savefig("no_headspace_CO2_O2.png")
```
